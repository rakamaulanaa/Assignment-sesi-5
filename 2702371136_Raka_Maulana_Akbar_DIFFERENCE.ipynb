```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)
# compute vector of forward differences
forward_diff = np.diff(y)/h
# compute corresponding grid
x_diff = x[:-1:]
# compute exact solution
exact_solution = -np.sin(x_diff)
# Plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
        label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution, \
          label = 'Exact solution')
plt.legend()
plt.show()
# Compute max error between
# numerical derivative and exact solution
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.1

# Define grid
x = np.array([0, 0.1, 0.2, 0.3, 0.4])
y = np.array([0.0000, 0.0819, 0.1341, 0.1646, 0.1791])

# Compute vector of forward differences
forward_diff = np.diff(y) / h

# Compute corresponding grid
x_diff = x[:-1]

# Plot solution
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite difference approximation')

plt.legend()
plt.xlabel('x')
plt.ylabel('Derivative')
plt.title('Forward Difference Approximation')
plt.grid(True)
plt.show()

# Compute derivatives at specific points
f_prime_at_0 = forward_diff[0]
f_prime_at_0_2 = forward_diff[2]

print(f"f'(0) ≈ {f_prime_at_0}")
print(f"f'(0.2) ≈ {f_prime_at_0_2}")
```
